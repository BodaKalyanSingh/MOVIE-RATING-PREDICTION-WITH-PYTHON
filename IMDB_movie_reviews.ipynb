```python
# Model Training and Testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score,accuracy_score
```

```python
data['Year'] = data['Year'].astype(str)
data['Duration'] = data['Duration'].astype(str)

data['Year'] = data['Year'].str.extract('(\d+)').astype(float)
data['Duration'] = data['Duration'].str.extract('(\d+)').astype(float)

X = data[['Duration', 'Year', 'Votes']]
Y = data[['Rating']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train['Votes'] = imputer.fit_transform(X_train[['Votes']])
```

```python
print(X_train.isnull().sum())
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
# Drop rows with missing values
X_test = X_test.dropna()
```

```python
X_test.isnull().sum()
```

```python
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
Y_pred=model.predict(X_test)
```

```python
print(np.unique(Y_test), np.unique(Y_pred))

# Convert to NumPy arrays if needed
Y_test = np.array(Y_test)
Y_pred = np.array(Y_pred)

# Find common indices
common_indices = np.where(Y_test == Y_pred)[0]

# Check if there are common indices
if len(common_indices) == 0:
    print("No common indices found.")
else:
    # Calculate metrics using common samples
    mae = mean_absolute_error(Y_test[common_indices], Y_pred[common_indices])
    rmse = mean_squared_error(Y_test[common_indices], Y_pred[common_indices])
    r2 = r2_score(Y_test[common_indices], Y_pred[common_indices])

    print(f'Mean absolute error: {mae}')
    print(f'Root mean squared error: {rmse}')
    print(f'R2 score: {r2}')
```
